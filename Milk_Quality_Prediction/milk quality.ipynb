### I convert the string values in the "Grade" column to numeric values.

```python
df.loc[df["Grade"] == 'high', "Grade"] = 2
df.loc[df["Grade"] == 'medium', "Grade"] = 1
df.loc[df["Grade"] == 'low', "Grade"] = 0
df.head()
```

```python
df.info()
```

```python
df["Grade"] = df["Grade"].astype(str).astype(int)
```

```python
df.info()
```

```python
df.corr()
```

```python
corr_matrix = df.corr()
corr_matrix["Grade"].sort_values(ascending=False)
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=np.bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, annot=True, mask=mask, cmap=cmap, vmax=.3, center=0,
 square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

# training algorithms

```python
x=df.iloc[:,:7]
x.head()
```

```python
y=df.iloc[:,7]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.3)
x_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier()
```

```python
KNN.fit(x_train, y_train)
```

```python
predictions =KNN.predict(x_test)
predictions
```

```python
KNN.score(x_train,y_train)
```

```python
KNN.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
resultKNN = classification_report(y_test, predictions)
print(resultKNN)
```

```python
output=KNN.predict([[6.6,1,1,1,1,1,240]])
output
```

```python
if output == 2:
    print("The milk quality is 'Good'")
if output == 1:
    print("The milk quality is 'Moderate'")
if output == 0:
    print("The milk quality is 'Bad'")
```
